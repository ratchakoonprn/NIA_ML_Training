```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42) #  seed  42
x = 10 * rng.rand(50) # x  50 
y = 2 * x - 1 + rng.randn(50) # y  50   (2* x - 1)  constant 
plt.scatter(x, y); 
```

```python
from sklearn.linear_model import LinearRegression # model  Linear Regression
model = LinearRegression(fit_intercept=True) # model  (intercept)  model
model # model
```

```python
x # x 
```

```python
x.shape
```

```python
y
```

```python
y.shape
```

```python
X = x[:, np.newaxis] # row vector  coloumn vector  X
X
```

```python
X.shape
```

```python
# features matrix  [n_samples, n_features]  x(50, 1)
# target array  [n_samples, n_targets] y(50,)
model.fit(X, y)
```

```python
model.coef_ # (coefficient)  model   feature
```

```python
model.intercept_ # model
```

```python
xfit = np.linspace(-1, 11) #  -1  11
xfit
```

```python
 # row vector  coloumn vector  features matrix  [n_samples, n_features]
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit) # y
```
